```python
from datetime import datetime

# Convert datetime string to datetime object
df['first_like_unlike_at'] = pd.to_datetime(df['first_like_unlike_at'], format='%Y-%m-%dT%H:%M:%S.%fZ')

# Get current datetime
current_datetime = datetime.now()

# Calculate timedelta and convert to hours
df['first_like_unlike_at'] = (current_datetime - df['first_like_unlike_at']) / pd.Timedelta('1 hour')
```
