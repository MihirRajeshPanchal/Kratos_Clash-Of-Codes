```python
disliked_people=[]
liked_people=[]
for i in range(df1.shape[0]):
    if df1['first_type'][i]=="DISLIKE":
        disliked_people.append(df1['p2'][i])
    if df1['first_type'][i]=="LIKE":
        liked_people.append(df1['p2'][i])
```

```python
print(disliked_people)
```

```python
from collections import Counter

my_disliked_people_dict = {key: count for key, count in Counter(disliked_people).items()}

print(my_disliked_people_dict)
```

```python
from collections import Counter

my_liked_people_dict = {key: count for key, count in Counter(liked_people).items()}

print(my_liked_people_dict)
```

```python
df2['liked_counts'] = df2['_id'].map(my_liked_people_dict)
```

```python
df2['disliked_counts'] = df2['_id'].map(my_disliked_people_dict)
```

```python
for i in range(df1.shape[0]):
    
    if df1['first_type'][i]=="DISLIKE":
        disliked_people.append(df1['p2'][i])
    if df1['first_type'][i]=="LIKE":
        liked_people.append(df1['p2'][i])
```

```python
like_score = (df2['liked_counts']/(df2['liked_counts']+df2['disliked_counts']))*100

height_above_score = df2[df2['height'] > 140]['height']
# height_above_score

height_below_score = df2[df2['height'] <= 140]['height']
# height_below_score

verified_score = df2[df2['is_verified'] == True]['is_verified']
# verified_score

unverified_score = df2[df2['is_verified'] != True]['is_verified']
# unverified_score

subscribed_score = df2[df2['is_subscribed'] == True]['is_subscribed']
# subscribed_score

unsubscribed_score = df2[df2['is_subscribed'] != True]['is_subscribed']
# unsubscribed_score



```

```python
# df2['Score']=(df2['liked_counts']/(df2['liked_counts']+df2['disliked_counts']))*100
df2['Score'] = (like_score + height_above_score + verified_score + subscribed_score) / 10 
```
