```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("users.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
from datetime import datetime

# Convert datetime string to datetime object
df['createdAt'] = pd.to_datetime(df['createdAt'], format='%Y-%m-%dT%H:%M:%S.%fZ')

# Get current datetime
current_datetime = datetime.now()

# Calculate timedelta and convert to hours
df['createdAt'] = (current_datetime - df['createdAt']) / pd.Timedelta('1 hour')
```

```python
df['dob'] = df['dob'].astype('object')
```

```python
from datetime import datetime

# Convert 'dob' column to datetime
df['dob'] = pd.to_datetime(df['dob'], format='%Y-%m-%dT%H:%M:%S.%fZ')

# Calculate age from 'dob' column in years
current_year = pd.Timestamp.now().year
df['age'] = current_year - df['dob'].dt.year

```

```python
from datetime import datetime

# Convert datetime string to datetime object
df['updatedAt'] = pd.to_datetime(df['updatedAt'], format='%Y-%m-%dT%H:%M:%S.%fZ')

# Get current datetime
current_datetime = datetime.now()

# Calculate timedelta and convert to hours
df['updatedAt'] = (current_datetime - df['updatedAt']) / pd.Timedelta('1 day')
```

```python
from datetime import datetime

# Convert datetime string to datetime object
df['verified_at'] = pd.to_datetime(df['verified_at'], format='%Y-%m-%dT%H:%M:%S.%fZ')

# Get current datetime
current_datetime = datetime.now()

# Calculate timedelta and convert to hours
df['verified_at'] = (current_datetime - df['verified_at']) / pd.Timedelta('1 day')
```
